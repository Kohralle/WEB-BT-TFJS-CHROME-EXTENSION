```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data1 = pd.read_csv('export1.csv')
```

```python
xs1 = data1['x_axis']
ys1 = data1['y_axis']
zs1 = data1['z_axis']
```

```python
fig = plt.figure(figsize=(30, 4))

xaxis = plt.plot(xs1, 'r',  label='x-axis')
yaxis = plt.plot(ys1, 'b', label='y-axis')
zaxis = plt.plot(zs1, 'g', label='z-axis')

plt.legend(loc = 3, fontsize='x-large')

plt.ylabel('acceleromter values', fontsize=16)
plt.xlabel('Reading ID (order of occurrence)', fontsize=16)

plt.show()
```
