```python
import tensorflow as tf
from tensorflow import keras
```

```python
import json
with open('history.json', 'r') as f:
    data = json.load(f)
    print(data['epoch'])
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 4))
plt.plot(data['history']['acc'], 'r', label='Accuracy of training data')
plt.plot(data['history']['val_acc'], 'b', label='Accuracy of validation data')
plt.title('Model Accuracy and Loss')
plt.ylabel('Accuracy and Loss')
plt.xlabel('Training Epoch')
plt.ylim(0)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 4))
plt.plot(data['history']['loss'], 'r', label='Loss of training data')
plt.plot(data['history']['val_loss'], 'b', label='Loss of validation data')
plt.title('Model Accuracy and Loss')
plt.ylabel('Accuracy and Loss')
plt.xlabel('Training Epoch')
plt.ylim(0)
plt.legend()
plt.show()
```
